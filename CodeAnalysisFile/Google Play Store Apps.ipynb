## 30. Average Size of Apps by Category

```python
df_csv.columns
```

```python
df_csv['Size']
```

```python
df_csv['Category']
```

```python
def convert_size(size):
    size = size.replace(',', '').replace('+', '')
    if 'M' in size:
        return float(size.replace('M', '')) * 1_000_000
    elif 'k' in size:
        return float(size.replace('k', '')) * 1_000
    elif size == 'Varies with device':
        return float('nan')
    return float(size)
```

```python
df_csv['Size'] = df_csv['Size'].apply(convert_size)
```

```python
df_csv['Size']
```

```python
df_csv.groupby('Category')['Size'].mean().sort_values(ascending = False)
```

```python
#visualisation

avg_size_cat = df_csv.groupby('Category')['Size'].mean().reset_index()

plt.figure(figsize = (14, 8))
sns.barplot(x = 'Size', y = 'Category', data = avg_size_cat, palette = 'viridis')
plt.title('Average Size of Apps by Category')
plt.xlabel('Average Size (MB)')
plt.ylabel('Category')
plt.show()
```
